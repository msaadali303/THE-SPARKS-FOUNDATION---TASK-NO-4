```python
df.rename(columns={'iyear':'Year', 'imonth':'Month', 'iday':'Day', 'country_txt':'Country', 'region_txt':'Region',
                   'provstate':'State', 'city':'City', 'attacktype1_txt':'Attack_Type', 'targtype1_txt':'Target_Type',
                   'targsubtype1_txt':'Target_Subtype', 'gname':'Gang_Name', 'weapsubtype1_txt':'Weapon_Type'},inplace=True)
```

```python
df.head()
```

```python
Terrorism = df[['Year','Month','Day','City','Country','State','Region','Attack_Type','Target_Type','Target_Subtype',
                'Gang_Name','Weapon_Type']]
```

```python
Terrorism.head()
```

```python
Terrorism.isnull().sum()
```

```python
Terrorism['City'].fillna(0,inplace=True)
Terrorism['State'].fillna(0,inplace=True)
Terrorism['Target_Subtype'].fillna(0,inplace=True)
Terrorism['Weapon_Type'].fillna(0,inplace=True)
```

```python
plt.subplots(figsize = (15,8))
sns.countplot(x = 'Year', data = Terrorism, color = 'teal')
plt.xticks(x = 'Year', rotation = 'vertical')
plt.title('Number of Terrorist each Year')
plt.show()
```

```python
plt.subplots(figsize = (15,8))
sns.countplot(x = 'Region', data = Terrorism , order = Terrorism['Region'].value_counts().index, color = 'teal') 
plt.xticks(x = 'Region',rotation = 'vertical')
plt.title('Most Target Region')
plt.show()
```

```python
plt.subplots(figsize = (15,8))
sns.countplot(x = 'City', data=Terrorism, order = Terrorism['City'].value_counts().head(10).index, color = 'teal')
plt.xticks(x = 'City',rotation = 'vertical')
plt.title('Most Target City')
plt.show()
```

```python
plt.subplots(figsize = (15,8))
sns.countplot(x = 'Country', data = Terrorism, order = Terrorism['Country'].value_counts().head(10).index, color = 'teal')
plt.title('Most Target Country')
plt.show()
```

```python
plt.subplots(figsize = (15,8))
sns.countplot(x = 'State', data = Terrorism, order = Terrorism['State'].value_counts().head(10).index, color = 'teal')
plt.title('Most Target State')
plt.show()
```

```python
sns.countplot(y = 'Attack_Type', data = Terrorism, order = Terrorism['Attack_Type'].value_counts().index, color = 'teal')
plt.xticks(x = 'Attack_Type', rotation = 90)
plt.title('Most Attack Type')
plt.show()
```

```python
sns.countplot(y = 'Weapon_Type', data = Terrorism, order = Terrorism['Weapon_Type'].value_counts().head(10).index, color = 'teal')
plt.xticks(x = 'Weapon_Type', rotation = 'vertical')
plt.title('Most Used Weapon')
plt.show()
```

```python
plt.subplots(figsize = (15,8))
sns.countplot(x = 'Target_Type', data = Terrorism, order = Terrorism['Target_Type'].value_counts().head(10).index, color = 'teal')
plt.xticks(x = 'Target_Type', rotation = 'vertical')
plt.title('Most Target Type')
plt.show()
```

```python
plt.subplots(figsize = (15,8))
sns.countplot(x = 'Region', data = Terrorism, order = Terrorism['Region'].value_counts().head(10).index, hue='Attack_Type')
plt.xticks(x = 'Region', rotation = 'vertical')
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Region VS Attack Type')
plt.show()
```

```python
sns.countplot(y = 'Gang_Name', data = Terrorism, order = Terrorism['Gang_Name'].value_counts().head(10).index, color = 'teal')
plt.xticks(x = 'Gang_Name', rotation = 'vertical')
plt.title('Highest Group Terror Attack')
plt.show()
```

```python
plt.subplots(figsize = (7,7))
sns.countplot(y = 'Gang_Name', data = Terrorism, order = Terrorism['Gang_Name'].value_counts().head(10).index, hue='Attack_Type')
plt.xticks(x = 'Gang_Name', rotation = 'vertical')
plt.xlabel('Count')
plt.ylabel('Gang Name')
plt.title('Gang VS Attack Type')
plt.show()
```
